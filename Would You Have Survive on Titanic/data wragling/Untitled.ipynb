```python
import pandas as pd
import numpy as np
```

```python
genders = pd.read_csv('gender_submission.csv')
genders.head()
```

```python
passengers = pd.read_csv('test.csv')
passengers.head()
```

```python
passengers['Sex'] = passengers['Sex'].apply(lambda x: 1 if x == 'female' else 0)
passengers.head()
```

```python
passengers['Age'].isnull().sum()
```

```python
passengers['Age'].fillna(value = np.mean(passengers['Age']), inplace = True)
passengers['Age'].isnull().sum()
```

```python
passengers['FirstClass'] = passengers['Pclass'].apply(lambda x: 1 if x == 1 else 0)
passengers['SecondClass'] = passengers['Pclass'].apply(lambda x: 1 if x==2 else 0)
passengers.head()
```

```python
passengers = passengers.merge(genders, on = 'PassengerId')
```

```python
passengers = passengers[['Sex', 'Age', 'FirstClass', 'SecondClass', 'Survived']]
passengers.head(50)
```

```python
passengers.isnull().sum()
```

```python
passengers.to_csv('./../passengers.csv')
```
